```python
# Funzione per trasformare le colonne categoriche finora salvate con float e int in stringhe, per evitare di passare
# informazioni inesistenti alla macchina
def cast_columns(df: pd.DataFrame, columns: list) -> pd.DataFrame:
    for col in columns:
        df[col] = df[col].astype(str)
    return df


# Nei casi in cui ho la data codificata in un'unica colonna in formato datetime, questa funzione mi restituisce 4 colonne
# gestibli dalle funzioni di training e dello stesso formato degli altri dataset
def datesplitter(df):
    
    # Estraggo mese, giorno e ora
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['hour'] = df['date'].dt.hour
    
    # Estraggo giorno della settimana in formato 3 lettere
    df['week'] = df['date'].dt.strftime('%a')
    # Imposto la prima lettera come lowercase per compatibilità con gli altri dataset
    df['week'] = df['week'].str[0].str.lower() + df['week'].str[1:]
    
    # Elimino la colonna della data
    df.drop(columns='date', inplace=True)


# Funzione "giocattolo" per pescare UN elemento a caso dell'insieme test (in principio potrei anche usare quello di 
# training) e vedere come va il modello su quel punto
def prova_test(modello, insieme_test, margin):
    
    # Seleziono il punto a caso
    indice = random.randint(0, len(insieme_test))
    
    # Inserisco il punto nel modello, ottenendo una predizione.
    Predizione = modello.predict(insieme_test[insieme_test.index==indice].drop(columns='CONSUMI'))[0]
    
    # Pesco il avlore reale registrato
    Fattuale   = insieme_test[insieme_test.index==indice].CONSUMI.values[0]
    
    # Trovo il discostamento in %
    rapporto = (Predizione/Fattuale-1)*100
    
    # Stampo i valori
    print("Il valore predetto è "+f'{Predizione:.3f}'+" W\nIl valore reale    è "+f'{Fattuale:.3f}'+" W\n")
    
    # Valuto se è accettabile o meno in base al margine scelto a priori
    if rapporto>0:
        if abs(rapporto)<margin:
            print("Si discostano del "+Fore.GREEN +"+"+ f'{rapporto:.1f}' +"%"+ Style.RESET_ALL + ".\n")
            print("Questa è una previsone accettabile")
        elif abs(rapporto)>margin:
            print("Si discostano del "+Fore.RED +"+"+ f'{rapporto:.1f}'+"%"+ Style.RESET_ALL+".\n")
            print("Questa " +Fore.RED + "NON" + Style.RESET_ALL+ " è una previsione accettabile")
    elif rapporto<0:
        if abs(rapporto)<margin:
            print("Si discostano del "+Fore.GREEN + f'{rapporto:.1f}' +"%"+ Style.RESET_ALL + ".\n")
            print("Questa è una previsone accettabile")
        elif abs(rapporto)>margin:
            print("Si discostano del "+Fore.RED + f'{rapporto:.1f}'+"%"+ Style.RESET_ALL+".\n")
            print("Questa " + Fore.RED + "NON" + Style.RESET_ALL+ " è una previsione accettabile")



# Versione più seria della funzione precedente, che opera su tutto l'iniseme passato e costruisce un grafico che mostra la
# percentuale di risultati buoni secondo il margine scelto sopra, quelli non buoni e quelli veramente terribili (sbagliati
# di almeno il 100%).
#
# GLi argomenti sono un modello o pipeline, un DataFrame del formato usato per allenare il modello/pipeline contente anche
# il target, una str con il nome della colonna target, un int/foat con il margine scelto sopra, una str con il nome del
# modello da usare nel titolo del grafico in uscita
def plot_percentage_error(model, df_test: pd.DataFrame, target: str, margin, model_name: str):
    # Estraggo la variabile target
    y_test = df_test[target]
    
    # Elimino il target dall'insieme su cui fare le predizioni
    X_test = df_test.drop(columns=target)
    
    # Uso il modello per fare le predizioni
    y_pred = model.predict(X_test)
    
    # Calcolo l'errore relativo in valore assoluto, con delle accortezze in caso di diviosni per 0
    with np.errstate(divide='ignore', invalid='ignore'):
        errors = abs(y_pred / y_test - 1)*100
        errors = pd.Series(errors)
        errors[~np.isfinite(errors)] = np.nan
    
    # Categorizzo gli errori: 
    # - se meno del margine lo considero ACCETTABILE;
    # - se più del 100% di errore, lo considero MOLTO SBAGLIATO;
    # - i rimanenti sono semplicemente NON ACCETTABILI.
    categories = errors.apply(lambda x: 'Accettabile' if x < margin or abs(x) < 1 else ('Molto sbagliato' if x > 100 else 'Non accettabile'))
    
    # Plotto l'istogramma delle categorie in %
    weights = np.ones(len(categories)) / len(categories)
    n, bins, patches = plt.hist(categories, weights=weights, rwidth=0.9, bins=3)
    
    # Aggiungo il valore del bin sul grafico
    for i in range(len(n)):
        plt.text(bins[i]+0.5*(bins[i+1]-bins[i]), n[i]+0.01, f"{n[i]:.2%}", ha='center')
    
    # Formattazione del grafico
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.grid(axis='y')
    plt.ylim(top=max(n)+0.1)
    plt.title(f'Istogramma delle categorie di errore ({model_name})')
    plt.xlabel('')
    plt.ylabel('Percentuale')
    plt.show()



import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import PercentFormatter

def plot_classification_error(model, df_test: pd.DataFrame, target: str, model_name: str):
    # Extract the target variable
    y_test = df_test[target]
    
    # Remove the target from the test data
    X_test = df_test.drop(columns=target)
    
    # Use the model to make predictions
    y_pred = model.predict(X_test)
    
    # Calculate the classification errors
    errors = y_pred != y_test
    
    # Calculate the percentage of correct and incorrect classifications
    correct = (errors == False).mean()
    incorrect = (errors == True).mean()
    
    # Plot the bar chart of correct and incorrect classifications
    plt.bar(['Correct', 'Incorrect'], [correct, incorrect])
    
    # Format the plot
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.grid(axis='y')
    plt.ylim(top=max(correct, incorrect)+0.1)
    plt.title(f'Classification error bar chart ({model_name})')
    plt.xlabel('')
    plt.ylabel('Percentage')
    plt.show()




    

# Funzione per implementare l'algoritmo random forest al DatFrame, accettando una lista di str coi nomi delle colonne
# categoriche
def fit_random_forest(df: pd.DataFrame, target: str, categorical: list, df_test_filename: str, df_train_filename:str) -> RandomForestRegressor:
    # Seleziono le feature escludendo il target
    features = [col for col in df.columns if col != target]
    
    # Uso One-hot encoding sulle feature categoriche
    df = pd.get_dummies(df, columns = categorical)
    features = [col for col in df.columns if col != target]
    
    # Inserisco valori mancanti usando un inserimento di valori medi
    imputer = SimpleImputer(strategy='mean')
    df[features] = imputer.fit_transform(df[features])
    
    # Divido gli insiemi di training e di test; uso random state 0 per ragionidi riproducibilità.
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)
    
    # Unisco il target e i dati dell'insieme di test in un unico dataframe per poterlo salvare
    df_test = pd.concat([X_test, y_test], axis=1)
    
    # Salvo il dataframe in un file .pkl
    df_test.to_pickle(df_test_filename)
    
    # Faccio lo stesso con l'insieme di allenamento
    df_train = pd.concat([X_train, y_train], axis=1)
    df_train.to_pickle(df_train_filename)
    
    # Uso tutte le risorse del sistema: mi faccio dare il numero di threads disponibli
    n = os.cpu_count()
    
    # Creo e fitto il modello Random Forest, usando 100 estimatori (scelto a caso) e sfruttando la parallelizzazione
    rf = RandomForestRegressor(n_estimators=100, random_state=0, n_jobs=n)
    rf.fit(X_train, y_train)
    
    # Uso l'insieme di test per valutare il modello
    y_pred = rf.predict(X_test)
    print('R^2:', r2_score(y_test, y_pred))
    print('MAE:', mean_absolute_error(y_test, y_pred))
    print('MSE:', mean_squared_error(y_test, y_pred))
    
    return rf


# Impementazione di un algoirtmo di neural netwoek; ho fatto molti test con reti di diverse dimensioni e parametri 
# [1, 2 e 3 layer; 4, 6, 10, 20, 50, 100, 200 e 500 neuroni per layer; regolarizzatore alpha da 0.1, 0.01, 0.001, 0.0001; 
# diversi tipi di attivatori; diversi metodi di risoluzione]  e questi sono quelli che hanno dato i risultati migliori.
# Nonostante questo, il modello risultante è terribile, soprattutto se paragonato alla Random Forest.
def train_nn_pipeline(df: pd.DataFrame, target: str, categorical: list, filename_train: str, filename_test: str) -> Pipeline:
    
    # Controlli per l'integrità dei dati
    if df.isnull().values.any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei NaN. Verranno rimpiazzati con le medie")
        df = df.fillna(df.median())
        
    numeric_cols = df.select_dtypes(include=np.number).columns
    
    if np.isinf(df[numeric_cols].values).any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei valori infiniti. Verranno rimpiazzati con il massimo valore finito.")
        df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
        df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].max())
        
    if (df[numeric_cols].values >= np.finfo(np.float64).max).any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei valori tropo grandi per float64. Verranno rimpiazzati con il valore massimo per float64.")
        df[numeric_cols] = df[numeric_cols].where(df[numeric_cols] < np.finfo(np.float64).max, np.finfo(np.float64).max)
    
    
    
    
    # Seleziono le features escludendo la colonna contenenteil target
    features = [col for col in df.columns if col != target]
    
    # Faccio lo splitting in insieme di training e di test
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)
    
    # Creo un preprocessore con un ColumnTransformer che applica il One Hot encoding alle colonne categoriche
    preprocessor = ColumnTransformer([
        ('onehot', OneHotEncoder(), categorical)
    ], remainder='passthrough')
    
    # Creo la pipeline con il preprocessore, uno scalatore e un regressore. Uso random_state fisato per ragioni di
    # riproducibilità. Il numero massimo di iterazioni standard è 200 e in molti casi non è sufficiente; in alcuni
    # casi anche 4000 si è rivelato insufficiente, quindi metto un nuemro finito ma esgeratamente alto per sicurezza.
    if categorical == ['numero_cir', 'TIPO_LOC', 'week']:
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('scaler', MaxAbsScaler()),
            ('nn', MLPRegressor(random_state=0, max_iter=300000))
        ])
    else:
        pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('scaler', MaxAbsScaler()),
        ('nn', MLPRegressor(random_state=0, max_iter=30000))
        ])
    
    # Rilevo il numero di threads a disposizone del sistema
    n_jobs = os.cpu_count()
    
    # Definisco gli iperparametri da usare nella crossvalidazione con la GridSearch. Qui ho lasciato solo quelli
    # che hanno dato i risultati migliori
    if night!= 1:
        param_grid = {
            'nn__hidden_layer_sizes': [(10,10)],
            'nn__activation': ['tanh'],
            'nn__solver': ['adam'],
            'nn__alpha': [0.001]
        }
    elif night==1:
                param_grid = {
            'nn__hidden_layer_sizes': [(10,), (20,), (50,), (100,), (10,10), (20, 20), (50, 50), (100, 100), (10, 10, 10)],
            'nn__activation': ['tanh'],
            'nn__solver': ['adam'],
            'nn__alpha': [0.001, 0.01, 0.0001]
        }
    # Creo un oggetto GridSearchCV per effettuare il tuning degli iperparametri sopra specificati. Attivo il 
    # multiprocessing usando n_jobs. Faccio stampare il numero di modelli e fit da fare.
    grid_search = GridSearchCV(pipeline, param_grid, n_jobs=n_jobs, verbose=1)
    
    # Applico il fit ai dati di tarining
    grid_search.fit(X_train, y_train)
    
    
    # Stampo la loss curve
    best_nn = grid_search.best_estimator_.named_steps['nn']
    plt.plot(best_nn.loss_curve_)
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()
    
    
    # Estraggo il modello con le migliori performance
    best_pipeline = grid_search.best_estimator_
    
    # Stampo il punteggio ottenuto
    score = best_pipeline.score(X_test, y_test)
    print(f'Test score: {score:.2f}')
    print(grid_search.best_params_)

    # Salvo in un file pickle l'insieme di training
    df_train = pd.concat([X_train.reset_index(drop=True), y_train.reset_index(drop=True)], axis=1)
    df_train.to_pickle(filename_train)
    
    # Salvo in un file pickle l'insieme test
    df_test = pd.concat([X_test.reset_index(drop=True), y_test.reset_index(drop=True)], axis=1)
    df_test.to_pickle(filename_test)
    
    return best_pipeline


# Implementazione di un regressore basato su una support vector machine; è molto rapido e potente, ma richiede assolutamente
# dei dataset di dimensioni molto ridotte, perché le richieste computazionali scalano come il quadrato del numero di punti.
# Non riesco nemmeno a usare il dataset di solo Trento, quindi ho usato dei dataset ragruppati ulteriormente.
def train_svr_pipeline(df: pd.DataFrame, target: str, categorical: list, filename_train: str, filename_test: str) -> Pipeline:
    
    # Controllo l'integrità dei dati e in caso li riparo
    if df.isnull().values.any():
        print("WARNING: Input contains NaN values. They will be replaced with medians.")
        df = df.fillna(df.median())
        
    numeric_cols = df.select_dtypes(include=np.number).columns
    
    if np.isinf(df[numeric_cols].values).any():
        print("WARNING: Input contains infinite values. They will be replaced with the maximum finite value.")
        df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
        df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].max())
        
    if (df[numeric_cols].values >= np.finfo(np.float64).max).any():
        print("WARNING: Input contains values too large for float64. They will be replaced with the maximum value for float64.")
        df[numeric_cols] = df[numeric_cols].where(df[numeric_cols] < np.finfo(np.float64).max, np.finfo(np.float64).max)
    
    # Seleziono le features escludendo il bersaglio
    features = [col for col in df.columns if col != target]
    
    # Faccio lo split ta dati test e dati allenamento
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)
    
    # Creo un preprocessor dotato di un ColumnTransformer che applica la codifica One-Hot alle colonne categoriche
    preprocessor = ColumnTransformer([
        ('onehot', OneHotEncoder(), categorical)
    ], remainder='passthrough')
    
    # Creo una pipeline che usa il preprocessore, lo scaler e il regressore Support Vector Regressor.
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('scaler', MaxAbsScaler()),
#         ('svr', SVR(cache_size=20000)),   # tentativo di fornire più memoria ad ogni instance per velocizzare la 
                                            # regressione; fallito miseramente, lo rallenta.
        ('svr', SVR())
    ])
    
    # Definisco gli iperparametri su cui effettuare la cross-validation con il GridSearchCV; interruttore per conti irra-
    # -gionevoli
    if night==1:
        param_grid = {
            'svr__kernel': ['poly'],
            'svr__degree': [1, 2, 3, 4],
            'svr__gamma': ['scale'],
            'svr__coef0': [1],
            'svr__C': [10],
            'svr__epsilon': [0.1, 0.5]
        }
    else:
        param_grid = {
            'svr__kernel': ['poly'],
            'svr__degree': [4],
            'svr__gamma': ['scale'],
            'svr__coef0': [1],
            'svr__C': [10],
            'svr__epsilon': [0.5]
        }
    
    # Uso tutte le risorse del sistema: mi faccio dare il numero di threads disponibli
    n = os.cpu_count()
    
    # Creo un oggetto GridSearchCV per fare il tuning degli iperparametri specificati sopra
    grid_search = GridSearchCV(pipeline, param_grid, n_jobs=n, verbose=1)
    
    # Faccio il fit
    grid_search.fit(X_train, y_train)
    
    # Estraggo il modello con le migliori performance
    best_pipeline = grid_search.best_estimator_
    
    # Stampo gli iperparametri migliori e il punteggio del modello
    y_pred = best_pipeline.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f'Test MSE: {mse:.2f}')
    print(grid_search.best_params_)

    # Salvo l'insieme di train in un file
    df_train = pd.concat([X_train.reset_index(drop=True), y_train.reset_index(drop=True)], axis=1)
    df_train.to_pickle(filename_train)
    
    # Salvo l'insieme di test in un file
    df_test = pd.concat([X_test.reset_index(drop=True), y_test.reset_index(drop=True)], axis=1)
    df_test.to_pickle(filename_test)
    
    return best_pipeline


# Funzione di classificazione in "consumi alti" e "consumi bassi" in base alla località. Usa sia una Random Forest che una Support Vector 
# Machine, anche se questa viene disabilitata per dataset da più di 10 000 righe per ragioni di tempo.
def classify_consumption(df: pd.DataFrame, target: str, categorical: list, df_test_filename: str, df_train_filename:str, groupby_col: str):
    
    # Controlli per l'integrità dei dati: se ci sono celle problematiche, sostituisco con delle medie.
    if df.isnull().values.any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei NaN. Verranno rimpiazzati con le medie")
        df = df.fillna(df.median())
        
    numeric_cols = df.select_dtypes(include=np.number).columns
    
    if np.isinf(df[numeric_cols].values).any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei valori infiniti. Verranno rimpiazzati con il massimo valore finito.")
        df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
        df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].max())
        
    if (df[numeric_cols].values >= np.finfo(np.float64).max).any():
        print(Fore.RED +"ATTENZIONE!"+ Style.RESET_ALL+": L'input contiene dei valori tropo grandi per float64. Verranno rimpiazzati con il valore massimo per float64.")
        df[numeric_cols] = df[numeric_cols].where(df[numeric_cols] < np.finfo(np.float64).max, np.finfo(np.float64).max)
   
    
    # Calcolo i consumi medi per ogni valore della colonna posizionale passata
    avg_consumption = df.groupby(groupby_col)[target].mean()

    # Definisco i criteri di classificazione in base a se il consumo è più alto o più basso del valore trovato al passaggio 
    # sopra
    def consumption_classification(row):
        groupby_val = row[groupby_col]
        consumption = row[target]
        if consumption < avg_consumption.loc[groupby_val]:
            return 'Low'
        else:
            return 'High'

    # Applico la funzione di classificazione al dataset e salvo i risultati nella nuova colonna 'classification'
    df['classification'] = df.apply(consumption_classification, axis=1)

    # Uso la codifica One-Hot sulle colonne categoriche
    df = pd.get_dummies(df, columns=categorical)
    target = 'classification'
    features = [col for col in df.columns if col not in [target, 'CONSUMI']]

    # Splitto insiemi di training e test
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)

    # Salvo l'insieme di test
    df_test = pd.concat([X_test, y_test], axis=1)
    df_test.to_pickle(df_test_filename)

    # Salvo l'insieme di training
    df_train = pd.concat([X_train, y_train], axis=1)
    df_train.to_pickle(df_train_filename)

    # Definisco la griglia degli iperparametri per la Random Forest
    rf_param_grid = {
        'n_estimators': [10, 50, 100],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'max_features': ['auto', 'sqrt']
    }

    # Faccio il tuning degli iperparametri con una GridSearch e una cross-validation a 5 fold
    rf_clf = RandomForestClassifier(random_state=0, n_jobs=-1)
    rf_grid_search = GridSearchCV(rf_clf, rf_param_grid, cv=5, n_jobs=-1)
    rf_grid_search.fit(X_train, y_train)

    # Alleno il classificatore usando i migliori iperparametri trovati sopra
    rf_clf = RandomForestClassifier(random_state=0, n_jobs=-1, **rf_grid_search.best_params_)
    rf_clf.fit(X_train, y_train)

    # Valuto il classificatore sull'insieme di test
    y_pred = rf_clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Random Forest test accuracy: {accuracy:.2f}')
    
    y_prob = rf_clf.predict_proba(X_test)
    rf_log_loss = log_loss(y_test, y_prob)
    print(f'Random Forest test log loss: {rf_log_loss:.2f}')

    # Stampo gli iperparametri usati per il Random Forest Classifier
    print(f'Random Forest best hyperparameters: {rf_grid_search.best_params_}')
    
    # Verifico il numero di righe del dataset per scegliere se allenare anche una SVM o se fermarmi qui
    if df.shape[0] < 10000:
        # Definisco due griglie di iperparametri da usare per il tuning della SVM; una grande per la notte, un altro con solo
        # il valore ottimale trovato durante i conti notturni
        if night==1:
            svm_param_grid = {
                'C': [0.1, 1, 10],
                'gamma': [0.1, 1, 10],
                'kernel': ['linear', 'rbf', 'poly'],
                'degree': [2, 3, 4]
            }
        else:
            param_grid = {
                'kernel': ['poly'],
                'degree': [4],
                'gamma': ['scale'],
                'coef0': [1],
                'C': [10],
                'epsilon': [0.5]
                }

        # Faccio il tuning degli iperparametri con una GridSearch e una cross-validation a 5 fold
        svm_clf = SVC(probability=True, random_state=0)
        svm_grid_search = GridSearchCV(svm_clf, svm_param_grid, cv=5, n_jobs=-1)
        svm_grid_search.fit(X_train, y_train)

        # Alleno il classificatore usando i migliori iperparametri trovati sopra
        svm_clf = SVC(probability=True, random_state=0, **svm_grid_search.best_params_)
        svm_clf.fit(X_train, y_train)

        # Valuto il classificatore sull'insieme di test
        y_pred = svm_clf.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f'SVM test accuracy: {accuracy:.2f}')
        
        y_prob = svm_clf.predict_proba(X_test)
        svm_log_loss = log_loss(y_test, y_prob)
        print(f'SVM test log loss: {svm_log_loss:.2f}')

        # Stampo gli iperparametri usati per lo SVM classifier
        print(f'SVM best hyperparameters: {svm_grid_search.best_params_}')
        
        return rf_clf, svm_clf
    else:
        print('Warning: Skipping SVM classifier because number of rows in input DataFrame is greater than or equal to 10,000')
        return rf_clf
```

# Funzione per la valutazione del numero di features

```python
def quante_features(df: pd.DataFrame, categories: list):
    print("Elementi unici nelle colonne categoriche:\n")

    features_number = 0
    for ii in categories:
        features_number = features_number + len(df[ii].unique())
        print("- " + ii + ": " + str(len(df[ii].unique())) + "\n")

    print("Queste dopo il One Hot encoding diventeranno features a loro volta.\n")

    print("Feature non categoriche: "+str(list(set(df.columns)-set(categories)))+", sono "+str(len(set(df.columns)-set(categories)))+"\n\n")

    features_number = features_number + len(set(df.columns)-set(categories))

    print("Numero di features totali: "+str(features_number))
```

```python
categorie = ['SEZ', 'TIPO_LOC', 'week']
quante_features(trient, categorie)
```
