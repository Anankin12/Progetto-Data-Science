```python
trento   = pd.read_pickle(files['trento'])

consumi = pd.read_pickle(files['consumi'])

grid   = pd.read_pickle(files['griglia'])

circoscrizioni = gpd.read_file(os.path.join(os.getcwd(), 'circoscrizioni.shp'))
circoscrizioni = circoscrizioni.to_crs('EPSG:4326')
circoscrizioni.drop(columns=['area', 'perimetro', 'fumetto'], inplace=True)

taa = gpd.read_file(os.path.join(os.getcwd(), 'R04_11_WGS84.shp'))
taa = taa.to_crs('EPSG:4326')
taa.drop(columns=['COD_ASC', 'COD_ZONA_C', 'COD_IS_AMM', 'COD_IS_LAC', 'COD_IS_MAR', 'COD_LAGUNA', 
                  'COD_VAL_P','LOC2011', 'COM_ASC', 'ACE', 'Shape_Leng', 'Shape_Area', 'COD_STAGNO', 
                  'COD_FIUME', 'COD_LAGO', 'COD_REG', 'COD_ISTAT', 'SEZ2011', 'COD_AREA_S', 
                  'COD_MONT_D'], inplace=True)

# taa.drop(columns=['COD_REG', 'COD_ISTAT', 'SEZ2011', 'COD_AREA_S', 'COD_MONT_D'], inplace=True)
tn = taa[taa.PRO_COM==22205].copy()
tn = tn.reset_index(drop=True).drop(columns = 'PRO_COM')
tn[['SEZ', 'TIPO_LOC', 'COD_LOC']] = tn[['SEZ', 'TIPO_LOC', 'COD_LOC']].astype(int)

# Create a spatial index on the circoscrizioni GeoDataFrame
circoscrizioni_sindex = circoscrizioni.sindex

# Perform a spatial join to find polygons that intersect with multiple borders
sjoin = gpd.sjoin(tn, circoscrizioni, how='inner', predicate='intersects')

# Define a function to calculate the overlap between each polygon and each administrative border
def calculate_overlap(row, borders):
    polygon = row.geometry
    # Use the cx indexer to perform a bounding box query on the borders GeoDataFrame
    possible_matches_index = list(borders.sindex.intersection(polygon.bounds))
    possible_matches = borders.iloc[possible_matches_index]
    precise_matches = possible_matches[possible_matches.intersects(polygon)]
    overlaps = []
    for j, border in precise_matches.iterrows():
        overlap = polygon.intersection(border.geometry).area
        overlaps.append((row.name, j, overlap))
    return overlaps

# Calculate the overlap between each polygon and each administrative border
overlaps = sjoin.apply(calculate_overlap, axis=1, args=(circoscrizioni,))
overlaps = pd.DataFrame([item for sublist in overlaps for item in sublist], columns=['polygon_index', 'border_index', 'overlap'])

# Find the border with the maximum overlap for each polygon
max_overlaps = overlaps.loc[overlaps.groupby('polygon_index')['overlap'].idxmax()]

# Merge the polygons and borders GeoDataFrames using the maximum overlap
result = pd.merge(max_overlaps, tn, left_on='polygon_index', right_index=True)
result = pd.merge(result, circoscrizioni, left_on='border_index', right_index=True)
result = result.reset_index(drop=True)
result.drop(columns=['polygon_index', 'border_index', 'overlap', 'COD_LOC', 'geometry_y'], inplace=True)
result = result.rename(columns={'geometry_x':'geometry'})
result = gpd.GeoDataFrame(result, geometry='geometry')
result[['SEZ', 'TIPO_LOC', 'numero_cir']] = result[['SEZ', 'TIPO_LOC', 'numero_cir']].astype(str)

trento[['SEZ', 'TIPO_LOC']]=trento[['SEZ', 'TIPO_LOC']].astype(str)

trento_geo = trento.merge(result, how='left', on=['SEZ', 'TIPO_LOC'])
trento_geo
trento[['SEZ', 'TIPO_LOC']]=trento[['SEZ', 'TIPO_LOC']].astype(int)
trento_geo = trento_geo.groupby(['month', 'day', 'week', 'hour', 
                                 'numero_cir', 'nome', 'TIPO_LOC'], as_index=False).agg({'NR_UBICAZIONI':'sum', 
                                                                                         'CONSUMI':'sum', 
                                                                                         'elevation':'mean', 
                                                                                         'TEMPERATURA':'mean'}).drop(columns=['nome'])
trento_geo[['numero_cir', 'TIPO_LOC']] = trento_geo[['numero_cir', 'TIPO_LOC']].astype(str)
trento_geo[['month', 'day', 'hour']] = trento_geo[['month', 'day', 'hour']].astype('int16')
def hour_interval(hour):
    if 0 <= hour <= 7:
        return 1
    elif 8 <= hour <= 15:
        return 2
    else:
        return 3

trento_geo['hour_interval'] = trento_geo['hour'].apply(hour_interval)
trento_geo.drop(columns='hour', inplace=True)
trento_geo = trento_geo.groupby(['month', 'day', 'week', 'hour_interval', 'numero_cir', 'TIPO_LOC', 'NR_UBICAZIONI', 'elevation'], as_index=False).agg({'CONSUMI':'sum', 'TEMPERATURA':'mean'})


dataset_completo = pd.read_pickle(files['totale'])

test3 = pd.read_pickle('test_SMALL.pkl')
test2 = pd.read_pickle('test_SEZ.pkl')
```

```python
def train_svr_pipeline(df: pd.DataFrame, target: str, categorical: list, filename_train: str, filename_test: str) -> Pipeline:
    
    # Data integrity checks
    if df.isnull().values.any():
        print("WARNING: Input contains NaN values. They will be replaced with medians.")
        df = df.fillna(df.median())
        
    numeric_cols = df.select_dtypes(include=np.number).columns
    
    if np.isinf(df[numeric_cols].values).any():
        print("WARNING: Input contains infinite values. They will be replaced with the maximum finite value.")
        df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
        df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].max())
        
    if (df[numeric_cols].values >= np.finfo(np.float64).max).any():
        print("WARNING: Input contains values too large for float64. They will be replaced with the maximum value for float64.")
        df[numeric_cols] = df[numeric_cols].where(df[numeric_cols] < np.finfo(np.float64).max, np.finfo(np.float64).max)
    
    # Select features excluding the target column
    features = [col for col in df.columns if col != target]
    
    # Split data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)
    
    # Create a preprocessor with a ColumnTransformer that applies one-hot encoding to categorical columns
    preprocessor = ColumnTransformer([
        ('onehot', OneHotEncoder(), categorical)
    ], remainder='passthrough')
    
    # Create a pipeline with the preprocessor, a scaler, and a regressor. Use a fixed random_state for reproducibility.
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('scaler', MaxAbsScaler()),
#         ('svr', SVR(cache_size=20000)),
        ('svr', SVR())
    ])
    if night==1:
        # Define hyperparameters to use in cross-validation with GridSearchCV
        param_grid = {
            'svr__kernel': ['poly'],
            'svr__degree': [1, 2, 3, 4],
            'svr__gamma': ['scale'],
            'svr__coef0': [1],
            'svr__C': [10],
            'svr__epsilon': [0.1, 0.5]
        }
    else:
        # Define hyperparameters to use in cross-validation with GridSearchCV
        param_grid = {
            'svr__kernel': ['poly'],
            'svr__degree': [4],
            'svr__gamma': ['scale'],
            'svr__coef0': [1],
            'svr__C': [10],
            'svr__epsilon': [0.5]
        }
    # Uso tutte le risorse del sistema: mi faccio dare il numero di threads disponibli
    n = os.cpu_count()
    
    # Create a GridSearchCV object to perform hyperparameter tuning using the specified hyperparameters.
    grid_search = GridSearchCV(pipeline, param_grid, n_jobs=n, verbose=1)
    
    # Fit the data to the training set
    grid_search.fit(X_train, y_train)
    
    # Extract the best model
    best_pipeline = grid_search.best_estimator_
    
    # Print the test score and best hyperparameters
    y_pred = best_pipeline.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f'Test MSE: {mse:.2f}')
    print(grid_search.best_params_)

    # Save training set to a pickle file
    df_train = pd.concat([X_train.reset_index(drop=True), y_train.reset_index(drop=True)], axis=1)
    df_train.to_pickle(filename_train)
    
    # Save test set to a pickle file
    df_test = pd.concat([X_test.reset_index(drop=True), y_test.reset_index(drop=True)], axis=1)
    df_test.to_pickle(filename_test)
    
    return best_pipeline
```

```python
import logging

logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

def train_rf(df, consumption_col, temperature_col, datetime_col, cat_cols, num_cols):
    logging.debug('Starting train_rf function')
    
    # Create lagged features for the past 3 days
    logging.debug('Creating lagged features')
    for i in range(1, 4):
        df[f'{consumption_col}_lag_{i}'] = df[consumption_col].shift(i)
        df[f'{temperature_col}_lag_{i}'] = df[temperature_col].shift(i)

    # Drop rows with missing values (first 3 rows)
    logging.debug('Dropping rows with missing values')
    df = df.dropna()

    # One-hot encode categorical columns
    logging.debug('One-hot encoding categorical columns')
    ohe = OneHotEncoder(sparse=False)
    cat_df = pd.DataFrame(ohe.fit_transform(df[cat_cols]), columns=ohe.get_feature_names_out(cat_cols), index=df.index)
    df = pd.concat([df, cat_df], axis=1)

    # Define features and target
    logging.debug('Defining features and target')
    features = [f'{consumption_col}_lag_1', f'{consumption_col}_lag_2', f'{consumption_col}_lag_3',
                f'{temperature_col}_lag_1', f'{temperature_col}_lag_2', f'{temperature_col}_lag_3']
    features += list(cat_df.columns) + num_cols
    target = consumption_col

    # Split data into training and test sets using train_test_split
    logging.debug('Splitting data into training and test sets')
    X = df[features]
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    # Join together X and y for training and test sets
    train_data = pd.concat([X_train, y_train], axis=1)
    test_data = pd.concat([X_test, y_test], axis=1)

    # Define hyperparameter grid for GridSearchCV
    logging.debug('Defining hyperparameter grid for GridSearchCV')
    param_grid = {
        'n_estimators': [10, 50, 100],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    # Perform hyperparameter tuning using GridSearchCV
    logging.debug('Performing hyperparameter tuning using GridSearchCV')
    rf = RandomForestRegressor(n_jobs=-1)
    grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
    grid_search.fit(X_train, y_train)

    # Print best hyperparameters
    print(f'Best hyperparameters: {grid_search.best_params_}')

    # Fit random forest regressor using best hyperparameters
    logging.debug('Fitting random forest regressor using best hyperparameters')
    rf = grid_search.best_estimator_
    rf.fit(X_train, y_train)

    # Make predictions on test set
    logging.debug('Making predictions on test set')
    y_pred = rf.predict(X_test)

    # Calculate and print R-squared, mean squared error, and mean absolute error scores on test data
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
```
